# ðŸ“ˆ Gradient Descent Implementation for LR

develop a Python program that is able to implement the gradient descent in order to achieve the linear regression of a set of datapoints.

### Imports

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
```

### LR Full Implementation 

```python
my_data = np.genfromtxt('RegData.csv', delimiter=',')
x = my_data[:,0]
y = my_data[:,1]
```

```python
theta0, theta1 = 0,0
m = len(x)
alpha = 0.001
epochs = 250
all_loss = []
all_theta0 = []
all_theta1 = []
```

```python
for i in range(epochs):
    y_hat  = theta0 + theta1 * x
    
    loss = sum( np.square(y_hat - y) ) / (2*m)
    
    all_loss.append(loss)
    all_theta0.append(theta0)
    all_theta1.append(theta1)
    
    # Gradient thetas
    g_theta0 = sum(y_hat - y) / m
    g_theta1 = sum((y_hat - y) * x) / m
    
    theta0 = theta0 - (alpha * g_theta0)
    theta1 = theta1 - (alpha * g_theta1)  
    
   
    

print(f"After {epochs} Iterations\n")
print(f"Theta0: {theta0}")
print(f"Theta1: {theta1}")
print(f"loss: {loss}\n")
print(f"Y_Hat: \n{y_hat}\n")
print(f"Y: \n{y}")
```

```python
# epochs - loss
plt.plot(np.arange(0,epochs), all_loss)
plt.title('Epochs - Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
# Using Matplotlib to plot a straight line. x, y_hat with previous scatter
plt.scatter(x,y)
plt.plot(x, y_hat)
plt.title('X - Y - Line Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# theta0 - loss
plt.plot(all_theta0, all_loss)
plt.title('Theta0 - Loss')
plt.xlabel('Theta0')
plt.ylabel('Loss')
plt.show()
```

```python
# theta1 - loss
plt.plot(all_theta1, all_loss)
plt.title('Theta1 - Loss')
plt.xlabel('Theta1')
plt.ylabel('Loss')
plt.show()
```

```python
# Accuracy
r2_score(y, y_hat)
```

### GD vectorize Implementation

```python
data = np.genfromtxt('MultipleLR.csv', delimiter=',')
x  = data[:,0:3]
x2 = np.ones(len(x))
y  = data[:,3]

X = np.insert(x, 0, x2, axis=1)
y = y.reshape(25,1)
```

```python
X
```

```python
Theta = np.zeros(4).reshape(4,1)
Theta
```

```python
def Batch_GD_Multi(x, y, alpha, epochs):
    m = len(X) 
    all_loss, all_thetas = [] , []
    Theta = np.zeros(4).reshape(4,1)
    
    for i in range(epochs):
        y_hat  = X@(Theta)
        
        loss = sum( np.square(y_hat - y) ) /(2*m)
        
        # Add previous thetas and loss 
        all_loss.append(loss)
        all_thetas.append(Theta)
        
        # Gradient thetas
        g_theta =  X.transpose().dot(y_hat - y) / m
        
        # Update thetas
        Theta = Theta - (alpha * g_theta)

        
        #Stopping using gradient
        if np.linalg.norm([g_theta])  < 0.01:
            break
        
        # Stopping using loss
        if (i > 0) and (all_loss[i-1] - all_loss[i]) < 0.001:
            break
        
        # Stopping using theta
        if (i > 0) and ( ( np.linalg.norm([all_theta0[i], all_theta1[i]]) 
                            - np.linalg.norm([all_theta0[i-1], all_theta1[i-1]])) < 0.0001):
            break
    
    print(f"After {i+1} Iterations\n")
    print(f"Theta: {Theta}")
    print(f"loss: {loss}\n")
    print(f"Y_Hat: \n{y_hat}\n")
    print(f"Y: \n{y}")
    
    return (all_thetas, all_loss, y_hat, loss, Theta, i)

all_thetas, all_loss, y_hat, loss, Theta, num_iterations \
= Batch_GD_Multi(X,y, 0.00001, 250)
```

```python
r2_score(y, y_hat)
```
