### Imports

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
import random
```

### Generate Data

```python
data = np.linspace(0, 20, num=50)
np.random.shuffle(data)
data
```

```python
x = data[:]
y = (-2 * x) + 1
```

## Batch Gradient Descent

```python
def Batch_GD(x, y, alpha, epochs):
    
    m = len(x) 
    all_loss, all_theta0, all_theta1 = [], [], []
    theta0, theta1 = 0, 0
    plot = plt.scatter(x, y, s=45)
    
    for i in range(epochs):
        y_hat  = theta0 + theta1 * x

        loss = sum( np.square(y_hat - y) ) / (2*m)
        
        # Add previous thetas and loss 
        all_loss.append(loss)
        all_theta0.append(theta0)
        all_theta1.append(theta1)
        
        # Gradient thetas
        g_theta0 = sum(y_hat - y) / m
        g_theta1 = sum((y_hat - y) * x) / m
        
        # Update thetas
        theta0 = theta0 - (alpha * g_theta0)
        theta1 = theta1 - (alpha * g_theta1)  
        
        plot = plt.plot(x, y_hat, alpha=0.5)
        plt.title('X - Y - Line Predicted on every iteration')
        plt.xlabel('X')
        plt.ylabel('y')
        
        # Stopping using gradient
        if np.linalg.norm([g_theta0, g_theta0]) < 0.001:
            break
        
        # Stopping using loss
        if (i > 0) and (all_loss[i-1] - all_loss[i]) < 0.0001:
            break
        
        # Stopping using theta
        if (i > 0) and ( ( np.linalg.norm([all_theta0[i], all_theta1[i]]) 
                            - np.linalg.norm([all_theta0[i-1], all_theta1[i-1]])) < 0.00001):
            break
    
    print(f"After {i+1} Iterations\n")
    print(f"Theta0: {theta0}")
    print(f"Theta1: {theta1}")
    print(f"loss: {loss}\n")
    print(f"Y_Hat: \n{y_hat}\n")
    print(f"Y: \n{y}")
    
    return (all_theta0, all_theta1, all_loss, y_hat, loss, theta0, theta1, i)

all_theta0, all_theta1, all_loss, y_hat, loss,theta0, theta1, num_iterations \
= Batch_GD(x,y, 0.001, 250)
```

```python
# Score
r2_score(y, y_hat)
```

```python
# epochs - loss
plt.plot(np.arange(0,num_iterations+1), all_loss)
plt.title('Epochs - Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
# Theta0 - Loss
plt.plot(all_theta0, all_loss)
plt.title('Theta0 - Loss')
plt.xlabel('Theta0')
plt.ylabel('Loss')
plt.show()
```

```python
# Theta1 - Loss
plt.plot(all_theta0, all_loss)
plt.title('Theta1 - Loss')
plt.xlabel('Theta1')
plt.ylabel('Loss')
plt.show()
```

```python
# Using Matplotlib to plot a straight line. x, y_hat with previous scatter
plt.scatter(x, y)
plt.plot(x, y_hat)
plt.title('The Best regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

## Mini Batch

```python
# Standard Solution
def Mini_Batch_2(x, y, alpha, epochs, b):
    
    m = len(x) 
    all_loss, all_theta0, all_theta1 = [], [], []
    theta0, theta1 = 0, 0
    
    final_list = lambda x, l: [x[i:i+l] for i in range(0, len(x), l)]
    # Choose b random samples from our exampels
    
    x_batches = final_list(x, b)
    y_batches = final_list(y, b)
    
    for i in range(epochs):
        for j in range(len(x_batches)):
            y_hat  = theta0 + theta1 * x_batches[j]

            loss = sum( np.square(y_hat - y_batches[j]) ) /(2*m)

            # Add previous thetas and loss 
            all_loss.append(loss)
            all_theta0.append(theta0)
            all_theta1.append(theta1)

            # Gradient thetas
            g_theta0 = sum(y_hat - y_batches[j]) / m
            g_theta1 = sum((y_hat - y_batches[j]) * x_batches[j]) / m

            # Update thetas
            theta0 = theta0 - (alpha * g_theta0)
            theta1 = theta1 - (alpha * g_theta1)  

            plot = plt.scatter(x_batches[j], y_batches[j])
            plot = plt.plot(x_batches[j], y_hat)
            plt.title('X - Y - Line Predicted on every iteration')
            plt.xlabel('X')
            plt.ylabel('y')
        
        y_hat = theta0 + theta1 * x
        
        # Stopping using gradient
        if np.linalg.norm([g_theta0, g_theta0]) < 0.0004:
            break
            
#         # Stopping using loss
#         if (i > 0) and (all_loss[i-1] - all_loss[i]) < 0.00001:
#             break
        
#         # Stopping using theta
#         if (i > 0) and ( ( np.linalg.norm([all_theta0[i], all_theta1[i]]) 
#                             - np.linalg.norm([all_theta0[i-1], all_theta1[i-1]])) < 0.0001):
#             break
    
    print(f"After {i+1} Iterations\n")
    print(f"Theta0: {theta0}")
    print(f"Theta1: {theta1}")
    print(f"loss: {loss}\n")
    print(f"Y_Hat: \n{y_hat}\n")
    print(f"Y: \n{y}")
    
    return (all_theta0, all_theta1, all_loss, y_hat, loss, theta0, theta1, i)

all_theta0, all_theta1, all_loss, y_hat, loss,theta0, theta1, num_iterations \
= Mini_Batch_2(x,y, 0.001, 250, 5)
```

```python
# Another Solution -> Just take random sample at a time
def Mini_Batch(x, y, alpha, epochs, b):
    
    m = len(x) 
    all_loss, all_theta0, all_theta1 = [], [], []
    theta0, theta1 = 0, 0
    
    for i in range(epochs):
        # Choose b random samples from our exampels
        random_sample = random.sample(range(0, 50), b)
        
        y_hat  = theta0 + theta1 * x[random_sample]

        loss = sum( np.square(y_hat - y[random_sample]) ) /(2*m)
        
        # Add previous thetas and loss 
        all_loss.append(loss)
        all_theta0.append(theta0)
        all_theta1.append(theta1)
        
        # Gradient thetas
        g_theta0 = sum(y_hat - y[random_sample]) / m
        g_theta1 = sum((y_hat - y[random_sample]) * x[random_sample]) / m
        
        # Update thetas
        theta0 = theta0 - (alpha * g_theta0)
        theta1 = theta1 - (alpha * g_theta1)  
        
        plot = plt.scatter(x[random_sample], y[random_sample])
        plot = plt.plot(x[random_sample], y_hat)
        plt.title('X - Y - Line Predicted on every iteration')
        plt.xlabel('X')
        plt.ylabel('y')
        
        # Stopping using gradient
        if np.linalg.norm([g_theta0, g_theta0]) < 0.01:
            break
    
    print(f"After {i+1} Iterations\n")
    print(f"Theta0: {theta0}")
    print(f"Theta1: {theta1}")
    print(f"loss: {loss}\n")
    print(f"Y_Hat: \n{y_hat}\n")
    print(f"Y: \n{y}")
    
    return (all_theta0, all_theta1, all_loss, y_hat, loss, theta0, theta1, i)

all_theta0, all_theta1, all_loss, y_hat, loss,theta0, theta1, num_iterations \
= Mini_Batch(x,y, 0.001, 250, 5)
```

## Stochastic GD 

```python
# Our Solution
def Stochastic(x, y, alpha, epochs, b):
    
    m = len(x) 
    all_loss, all_theta0, all_theta1 = [], [], []
    theta0, theta1 = 0, 0
    
    for i in range(epochs):
        for j in range(len(x)):
            y_hat  = theta0 + theta1 * x[j]

            loss = np.square(y_hat - y[j]) / 2

            # Add previous thetas and loss 
            all_loss.append(loss)
            all_theta0.append(theta0)
            all_theta1.append(theta1)

            # Gradient thetas
            g_theta0 = y_hat - y[j]
            g_theta1 = (y_hat - y[j]) * x[j]

            # Update thetas
            theta0 = theta0 - (alpha * g_theta0)
            theta1 = theta1 - (alpha * g_theta1)  
            
        y_hat =  theta0 + theta1 * x
        
        # Stopping using gradient
        if np.linalg.norm([g_theta0, g_theta0]) < 0.01:
            break
    
    print(f"After {i+1} Iterations\n")
    print(f"Theta0: {theta0}")
    print(f"Theta1: {theta1}")
    print(f"loss: {loss}\n")
    print(f"Y_Hat: \n{y_hat}\n")
    print(f"Y: \n{y}")
    
    return (all_theta0, all_theta1, all_loss, y_hat, loss, theta0, theta1, i)

all_theta0, all_theta1, all_loss, y_hat, loss, theta0, theta1, num_iterations \
= Stochastic(x,y, 0.0001, 100, 5)
```
